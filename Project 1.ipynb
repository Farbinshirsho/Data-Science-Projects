# Data Wrangling 

```python
# Loading the datasets
df_emissions = pd.read_csv("../Datasets/greenhouse-gas-emissions-industry-and-household-september-2025-quarter.csv")
df_gdp = pd.read_csv("../Datasets/regional-gross-domestic-product-year-ended-march-2025.csv")
```

```python
df_gdp.head(5)
```

```python
df_emissions.head(5)
```

```python
#Renaming the column names
```

```python
df_gdp = df_gdp.rename(columns={
    'Series_title_2': 'Region',
    'Series_title_3': 'Industry',
    'Data_value': 'GDP_Value'
})
df_gdp.head(5)
```

```python
#Checking on uniqe industry
```

```python
df_gdp['Industry'].unique()
```

```python
#Handling Unwanted data
```

```python
summary_rows = [
    'Total All Industries', 
    'Gross Domestic Product', 
    'GST on Production, Import Duties and Other Taxes'
]
```

```python
df_gdp_clean = df_gdp[~df_gdp['Industry'].isin(summary_rows)].copy()
df_gdp_clean = df_gdp_clean.dropna(subset=['Industry', 'GDP_Value'])
```

```python
df_gdp_clean.head(5)
```

```python
# Extracting the year
def extract_year(period_val):
    return int(str(period_val).split('.')[0])
```

```python
df_gdp_clean['Year'] = df_gdp_clean['Period'].apply(extract_year)
```

```python
df_gdp_clean[['Year', 'Region', 'Industry', 'GDP_Value']].head(5)
```

```python
df_gdp_clean['GDP_Millions'] = df_gdp_clean['GDP_Value']
```

```python
df_gdp_final = df_gdp_clean[['Year', 'Region', 'Industry', 'GDP_Millions']].copy()
```

```python
df_gdp_final.head(5)
```

```python
#Converting regions to NZ total for each sector for each year
```

```python
df_gdp_final = df_gdp_final[df_gdp_final['Region'] == 'New Zealand'].copy().reset_index(drop=True)
```

```python
df_gdp_final = df_gdp_final.drop(columns=['Region'])
```

```python
df_gdp_final.head(5)
```

```python
df_emissions.head(5)
```

```python
df_em_clean = df_emissions[
    (df_emissions['Gas'] == 'Carbon dioxide equivalents') & 
    (df_emissions['Variable'] == 'Actual')
].copy()
```

```python
#Extracting Year
```

```python
def extract_year(period_val):
    return int(str(period_val).split('.')[0])
    
df_em_clean['Year'] = df_em_clean['Period'].apply(extract_year)
```

```python
df_em_annual = df_em_clean.groupby(['Year', 'Anzsic_descriptor'])['Data_value'].sum().reset_index()
```

```python
df_em_annual.columns = ['Year', 'Industry', 'Emissions_Kilotonnes']
```

```python
print("Cleaned & Annualized Emissions Data:")
df_em_annual.head(5)
```

```python
df_em_annual['Industry'].unique()
```

```python
#mapping dictionary for industry
```

```python
industry_map = {
    'Agriculture, forestry, and fishing': 'Agriculture',
    'Mining': 'Forestry, Fishing, and Mining',
    'Electricity generation and gas supply': 'Electricity, Gas, Water, and Waste services',
    'Water, sewerage, drainage, and waste services': 'Electricity, Gas, Water, and Waste services',
    'Food, beverage, and tobacco product manufacturing': 'Manufacturing',
    'Wood and paper products manufacturing and printing': 'Manufacturing',
    'Petroleum, chemical, polymer, and rubber product manufacturing': 'Manufacturing',
    'Non-metallic mineral and metallic product manufacturing': 'Manufacturing',
    'Other manufacturing nec': 'Manufacturing',
    'Transport, postal, and warehousing': 'Transport, Postal and Warehousing',
    'Retail trade': 'Retail Trade',
    'Wholesale trade': 'Wholesale Trade',
    'Accommodation': 'Accommodation',
    'Education and training': 'Education and Training',
    'Health care and social assistance': 'Health Care and Social Assistance',
    'Public administration and safety': 'Public Administration and Safety',
    'Professional, scientific, and technical services': 'Professional, Scientific, and Technical Services'
}
df_em_annual['Industry'] = df_em_annual['Industry'].replace(industry_map)
df_em_grouped = df_em_annual.groupby(['Year', 'Industry'])['Emissions_Kilotonnes'].sum().reset_index()
df_final = pd.merge(df_gdp_final, df_em_grouped, on=['Year', 'Industry'])

df_final['Carbon_Intensity'] = df_final['Emissions_Kilotonnes'] / df_final['GDP_Millions']
display(df_final.head(5))
```

# EDA/Data Visulisation


```python
latest = df_gdp_final['Year'].max()
df_latest = df_gdp_final[df_gdp_final['Year'] == latest]

industry_gdp = df_latest.groupby('Industry')['GDP_Millions'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
industry_gdp.plot(kind='barh', color='pink')

plt.title(f'Top 10 NZ Industries by GDP Contribution ({latest})', fontsize=14)
plt.xlabel('GDP (Millions $NZ)', fontsize=12)
plt.ylabel('Industrial Sector', fontsize=12)

plt.show()
```

```python
latest_yr = df_final['Year'].max()
df_latest = df_final[df_final['Year'] == latest_yr].sort_values('Carbon_Intensity', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=df_latest, x='Carbon_Intensity', y='Industry', palette='magma')

plt.title(f'New Zealand Industrial Carbon Intensity ({latest_yr})', fontsize=14)
plt.xlabel('Intensity (Emissions per $1M GDP)')
plt.show()
```

```python
latest_yr = df_final['Year'].max()
df_summary = df_final[df_final['Year'] == latest_yr].groupby('Industry').agg({
    'GDP_Millions': 'sum',
    'Emissions_Kilotonnes': 'sum'
}).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 8))

sns.barplot(data=df_summary.sort_values('GDP_Millions', ascending=False).head(10), 
            x='GDP_Millions', y='Industry', color='green', ax=ax1, alpha=0.6, label='GDP')

ax2 = ax1.twiny()
sns.scatterplot(data=df_summary.sort_values('GDP_Millions', ascending=False).head(10), 
                x='Emissions_Kilotonnes', y='Industry', color='red', s=100, ax=ax2, label='Emissions')

ax1.set_title(f'Economic Contribution vs. Environmental Impact ({latest_yr})', fontsize=15)
ax1.set_xlabel('GDP ($NZ Millions) - Bars')
ax2.set_xlabel('Emissions (Kilotonnes CO2-e) - Red Dots')

plt.tight_layout()
plt.show()
```

```python
national_efficiency = df_final.groupby('Year')['Carbon_Intensity'].mean()

plt.figure(figsize=(10, 5))
plt.plot(national_efficiency.index, national_efficiency.values, marker='o',  color='forestgreen', linewidth=2)

plt.title('NZ Economic Carbon Intensity Trend (2010-2024)', fontsize=14)
plt.ylabel('Kilotonnes CO2-e per $1M GDP')
plt.xlabel('Year')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
latest_year = df_final['Year'].max()
df_scatter = df_final[df_final['Year'] == latest_year].groupby('Industry').agg({
    'GDP_Millions': 'sum',
    'Emissions_Kilotonnes': 'sum'
}).reset_index()

# Creating the Plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_scatter, x='GDP_Millions', y='Emissions_Kilotonnes', 
                hue='Industry', s=200, palette='viridis', legend=True)

#Adding Labels to each point (so you can see which dot is which)
for i in range(df_scatter.shape[0]):
    plt.text(df_scatter.GDP_Millions[i]+500, df_scatter.Emissions_Kilotonnes[i], 
             df_scatter.Industry[i], fontsize=9, alpha=0.8)

# Formatting
plt.title(f'Economic Value vs. Environmental Impact by Industry ({latest_year})', fontsize=15, fontweight='bold')
plt.xlabel('Economic Contribution (GDP $NZ Millions)', fontsize=12)
plt.ylabel('Greenhouse Gas Emissions (Kilotonnes CO2-e)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Industries')

plt.tight_layout()
plt.show()
```
